```python
import matplotlib.pyplot as plt

%matplotlib inline 

#by using this u can avoid using plt.show often often
```

```python
import numpy as np
```

```python
#simple examples 

x=np.arange(0,10)
y=np.arange(50,60)
```

```python
#ploting using matplotlib

#plt scatter

plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('test.png')
```

```python
y=x*x
```

```python
#plt plot

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('2D Graph')
plt.plot(y,'g-*')
```

```python
plt.plot(x,y,'g*',linestyle='dashed',linewidth=3,markersize=12)
```

```python
#creating sub plots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*--')
plt.subplot(2,2,4)
plt.plot(x,y,'g--')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title("MatplotLib demo")
plt.xlabel("X axis Caption")
plt.ylabel("Y axis Caption")
plt.plot(x,y,'r*-')
plt.show()
```

```python
#compute the x and y coordinates for points on a sine curve

x = np.arange(0, 4*np.pi,0.3)
y = np.sin(x)
plt.title('Sine wave form')

#plot the points using the matplotlib

plt.plot(x,y)
```

```python
#subplot()
#for sin and cosine curve forms 

x = np.arange(0, 5*np.pi,0.3)
y_sin = np.sin(x)
y_cos = np.cos(x)

#set up a subject frid that has height 2 and width 1,
#and set the first subplot such as sctive

plt.subplot(2,1,1)

#make the first plot

plt.plot(x,y_sin,'b*')
plt.title('cosine curve')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('sin curve')
plt.show()
```

# Bar Graph

```python
x = [2,8,9]
y = [6,2,9]

x2 = [4,3,1]
y2 = [4,3,8]
plt.bar(x, y)
plt.bar(x2, y2, color= 'g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

# Histogram

```python
a = np.array([22,12,33,34,11,32,55,64,76,87,88,44,34,34,54,27,90,200])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

#rectangular box plot

plt.boxplot(data,vert=True, patch_artist=True);
```

# Pie chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['red', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0) #expand 1st slice

#plot pie chart

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
